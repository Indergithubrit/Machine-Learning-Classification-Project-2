## Now preparing an  model on classification:

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=102)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
obj13=KNeighborsClassifier()
```

```python
obj13.fit(x_train,y_train)
```

## seeing the difference between training data and training data prediction:

```python
y_train
```

```python
x_train.values
```

```python
train_pred=obj13.predict(x_train.values)
train_pred
```

## Accuracy score and confusion matrix of training data:

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_train,train_pred)
```

```python
accuracy_score(y_train,train_pred)
```

## seeing the difference between testing data and testing data prediction:

```python
y_test
```

```python
x_test.values
```

```python
test_pred=obj13.predict(x_test.values)
test_pred
```

```python
confusion_matrix(y_test,test_pred)
```
